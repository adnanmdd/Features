#### Laptop Prediction using One-hot encoding

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import OneHotEncoder

from sklearn.model_selection import train_test_split
from sklearn.linear_model import Perceptron
from sklearn.metrics import accuracy_score
```

```python
# Dataset
dataset = {
    'Age':[22,35,28,40,23,31,26,38],
    'Preferred_OS':['Windows','macOS','Linux','Windows','Linux','macOS','Windows','Linux'],
    'Bought_Laptop':[1,1,0,1,0,1,0,1]
}

df = pd.DataFrame(dataset)
print(df)
```

```python
# One-hot encoding
df_encoded = pd.get_dummies(df,columns=['Preferred_OS'], dtype=int)
print('Encoded Dataset')

df_encoded
```

```python
# Features and Labels
X = df_encoded.drop('Bought_Laptop',axis=1)
y = df_encoded['Bought_Laptop']
```

```python
X = X.to_numpy()
y = y.to_numpy()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.25,random_state=42)
```

```python
# Perceptron Algo
model_encoded = Perceptron()
model_encoded.fit(X_train,y_train)

# Prediction
pred = model_encoded.predict(X_test)

# Accuracy
acc = accuracy_score(y_test,pred)
print("Accuracy of the model: ",acc)
```
