#### Predicting Rectangular Garden Area

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Dataset
data = {
    'Base_Adjustment(x)':[1,2,3,4,5,6],
    'Area(y)': [12,20,30,42,56,72]
}

df = pd.DataFrame(data)
df
```

```python
# Plot x vs y


X = df['Base_Adjustment(x)']
y = df['Area(y)']

plt.scatter(X,y,marker='x')
plt.title('Garden Area Growth')
plt.grid(True)
plt.xlabel('Base_Adjustment(x)')
plt.ylabel('Area(y)')
plt.show()
```

#####  **Note:** Area increases as x increases.

```python
# Create factored features

# Length - x+2, Width - x+3

length = X+2
width = X+3
factored_area = length*width

df_factored = pd.DataFrame({
    'x':X,
    'Length':length,
    'Width':width,
    'Area':factored_area
})

df_factored
```

```python
# Future Prediction
x_new = 20
future_area = (x_new+2)*(x_new+3)

print("Predicted Future Area: ",future_area)
```
